```python
# f1='Mg' 
# f2='Fe'   
# f3='RI'   
# f4='K'   
# f5 = 'Ca'   
# f6 = 'Si'     
# f7 = 'Na'     
# f8 = 'Ba'     
# f9 = 'Al'  

f1='Na'
f2 = 'Al'
f3 = 'Ba'

basex = np.array(xtrain[[f1,f2,f3]])
basey = np.array(ytrain["tar"])

testx = np.array(xtest[[f1,f2,f3]])
testy = np.array(ytest["tar"])
```

```python
def euclidean_distance(train_point,test_point):
    distance = np.sum((train_point - test_point)**2)
    
    return np.sqrt(distance)
```

```python
def calc_distance_from_all(all_points,given_point,predictions):
    all_distances = []
    
    for i, each in enumerate(all_points):
        distance = euclidean_distance(each,given_point)
        
        all_distances.append((distance,int(predictions[i])))
        
    all_distances.sort(key=lambda tup: tup[0])
    
    return all_distances
```

```python
def get_neighbours(distances,count):
    return distances[:count]                # prints distance,class pair of a pt
```

```python
# k is a value less than the square root of the total no. of rows of the dataset.
# The dataset has 214 rows whose square root is 14.62.
# but the no. of classes is 7 and k value should not be a multiple of no. of classes which is here 7 classes.
# let k =13

def predict(all_points,given_point,predictions,k):
    distances = calc_distance_from_all(all_points,given_point,predictions)
    neighbours = get_neighbours(distances,k)
    
    op = [row[-1] for row in neighbours]         # op = []
                                                  #  for row in neighbour row[-1] 
    predictions = max(set(op),key=op.count)    #set calculates the count of no. of unique values
    
    return predictions
```

```python
def accuracy(basex,basey,testx,testy,k):
    correct =0
    for i in range(len(testx)):
        p = predict(basex,testx[i],basey,k)
        
        if p == testy[i]:
            correct += 1
            
    return f"Accuracy: {correct*100/len(testy)}"
```

```python
k = 13              
accuracy(basex,basey,testx,testy,k)
```
