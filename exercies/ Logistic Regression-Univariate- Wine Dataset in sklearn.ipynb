```python
x = data['flavanoids']
y = data['TARGET']
```

```python
plt.plot(x,y,'r.')
```

```python
x = np.array(x)
y = np.array(y)
```

```python
x = (x-x.min())/(x.max()-x.min())
y = (y-y.min())/(y.max()-y.min())
```

```python
plt.plot(x,y,'r.')
```

```python
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size = 0.25)
```

```python
def hypothesis(a,x,b):
    return (1/(1+np.exp(-(a*x+b))))
```

```python
def error(a,x,b,y):
    e = 0
    m = len(x)
    for i in range(m):
        e += -(y[i]*np.log(hypothesis(a,x[i],b)))-((1-y[i])*np.log(1-hypothesis(a,x[i],b)))
        
    return (1/m)*e
```

```python
def step_gradient(a,x,b,y,learning_rate):
    grad_a = 0
    grad_b = 0
    m = len(x)
    
    for i in range(m):
        grad_a += (hypothesis(a,x[i],b)-y[i])*x[i]
        grad_b += (hypothesis(a,x[i],b)-y[i])
        
    a = a - (grad_a * learning_rate)
    b = b - (grad_b * learning_rate)
    
    return a,b
```

```python
def descend(init_a,x,init_b,y,learning_rate,iterations):
    a = init_a
    b = init_b
    
    for i in range(iterations):
        e = error(a,x,b,y)
        
        if i%1000 == 0:
            print(f"error: {e}, a:{a}, b:{b}")
            
        a,b = step_gradient(a,x,b,y,learning_rate)
        
    return a,b
```

```python
a =0
b=0
learning_rate = 0.01
iterations =10000
```

```python
final_a,final_b = descend(a,x_train,b,y_train,learning_rate,iterations)
```

```python
final_a,final_b
```

```python
error(final_a,x_train,final_b,y_train)
```

```python
error(final_a,x_test,final_b,y_test)
```

```python
plt.plot(x_train,y_train,'r.',
         x_train,hypothesis(final_a,x_train,final_b),'b.'
        )
```

```python
plt.plot(x_test,y_test,'r.',
        x_test,hypothesis(final_a,x_test,final_b),'b.'
        )
```

```python
def accuracy(a,x,b,y):
    correct = 0
    m = len(x)
    
    for i in range(m):
        pred = hypothesis(a,x[i],b)
        
        if pred > 0.5:
            z=1
        else:
            z = 0                 # 0 -benign
            
        if y[i] == z:
            correct += 1
            
    print(f"Accuracy: {correct/len(y)}")    
```

```python
accuracy(final_a,x_train,final_b,y_train)
```

```python
accuracy(final_a,x_test,final_b,y_test)
```
