```python
def hypothesis(a,b,c,d,x):
    return (a*np.power(x,3))+(b*np.power(x,2))+(c*x)+d
```

```python
def error(a,b,c,d,x,y):
    e = 0
    m = len(x)
    
    for i in range(m):
        e += np.power((hypothesis(a,b,c,d,x[i])-y[i]),2)
    return (1/(2*m))*e
```

```python
def step_gradient(a,b,c,d,x,y,learning_rate):
    grad_a = 0
    grad_b = 0
    grad_c = 0
    grad_d = 0
    m = len(x)
    
    for i in range(m):
        grad_a += (1/m)*(hypothesis(a,b,c,d,x[i]) - y[i])*(np.power(x[i],3))  #partial differentiation
        grad_b += (1/m)*(hypothesis(a,b,c,d,x[i]) - y[i])*(np.power(x[i],2))
        grad_c += (1/m)*(hypothesis(a,b,c,d,x[i]) - y[i])*x[i]
        grad_d += (1/m)*(hypothesis(a,b,c,d,x[i]) - y[i])
        
    a = a - grad_a * learning_rate
    b = b - grad_b * learning_rate
    c = c - grad_c * learning_rate
    d = d - grad_d * learning_rate
    
    return a,b,c,d
```

```python
def descend(init_a,init_b,init_c,init_d,x,y,learning_rate,iterations):
    a = init_a
    b = init_b
    c = init_c
    d = init_d
    
    for i in range(iterations):
        e = error(a,b,c,d,x,y)
        
        if i % 1000 ==0:
            print(f"Error: {np.sqrt(e)}, a:{a}, b:{b}, c:{c}, d:{d}")
            
        a,b,c,d = step_gradient(a,b,c,d,x,y,learning_rate)
    
    return a,b,c,d
```

```python
a = 0
b = 0
c = 0
d = 0
learning_rate = 0.001
iterations = 10000
```

```python
final_a,final_b,final_c,final_d = descend(a,b,c,d,x_train,y_train,learning_rate,iterations)
```

```python
plt.plot(x_train,y_train,'r.',
         x_train,hypothesis(a,b,c,d,x_train),'b.',
         x_train,hypothesis(final_a,final_b,final_c,final_d,x_train),'g.')
```
