```python
def hypothesis(a,x1,x2,b,c):
    return a*x1 + b*x2 + c
```

```python
def error(a,x1,x2,b,c,y):
    e = 0
    m = len(x1)
    
    for i in range(m):
        e += np.power((hypothesis(a,x1[i],x2[i],b,c)-y[i]),2)
    return (1/(2*m))*e
```

```python
def step_gradient(a,x1,x2,b,c,y,learning_rate):
    grad_a = 0
    grad_b = 0
    grad_c = 0
    m = len(x1)
    
    for i in range(m):
        grad_a += (1/m)*(hypothesis(a, x1[i],x2[i],b,c) - y[i])*x1[i]  #partial differentiation
        grad_b += (1/m)*(hypothesis(a, x1[i],x2[i],b,c) - y[i])*x2[i]
        grad_c += (1/m)*(hypothesis(a,x1[i],x2[i],b,c)-y[i])
        
    a = a - grad_a * learning_rate
    b = b - grad_b * learning_rate
    c = c - grad_c * learning_rate
    
    return a,b,c
```

```python
def descend(init_a,x1,x2,init_b,init_c,y,learning_rate,iterations):
    a = init_a
    b= init_b
    c = init_c
    
    for i in range(iterations):
        e = error(a,x1,x2,b,c,y)
        
        if i % 1000 == 0:
            print(f"Error: {np.sqrt(e)}, a:{a}, b:{b}, c:{c}")
            
        a,b,c = step_gradient(a,x1,x2,b,c,y,learning_rate)
    
    return a,b,c
```

```python
a = 0
b = 0
c = 0
learning_rate = 0.001
iterations = 10000
```

```python
final_a,final_b,final_c = descend(a,x1_train,x2_train,b,c,y_train,learning_rate,iterations)
```

```python
plt.plot(x1_train,x2_train,y_train,'r.',
         x1_train,x2_train,hypothesis(a,x1_train,x2_train,b,c),'b',
         x1_train,x2_train,hypothesis(final_a,x1_train,x2_train,final_b,final_c),'g')
```
