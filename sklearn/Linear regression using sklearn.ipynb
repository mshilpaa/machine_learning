```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

from sklearn.datasets import load_boston
```

```python
boston = load_boston()
```

```python
boston
```

```python
boston.data
```

```python
boston.feature_names
```

```python
features = pd.DataFrame(boston.data,columns = boston.feature_names)
features
```

```python
target = pd.DataFrame(boston.target,columns=['TARGET'])
target
```

```python
data = pd.concat([features,target],axis=1)
data
```

```python
x = np.array(data['RM'])
y = np.array(data['TARGET'])
```

```python
from sklearn.preprocessing import MinMaxScaler
```

```python
scaler = MinMaxScaler()
```

```python
x = x.reshape(-1,1)
y = y.reshape(-1,1)
```

```python
x.shape
```

```python
x = scaler.fit_transform(x)
y = scaler.fit_transform(y)
```

```python
plt.plot(x,y,'r.')
```

```python
from sklearn.model_selection import train_test_split 
```

```python
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.2)
```

```python
from sklearn.linear_model import LinearRegression
```

```python
regressor = LinearRegression() 
```

```python
regressor.fit(x_train,y_train)
```

```python
y_pred = regressor.predict(x_test)
```

```python
y_pred
```

```python
for i in zip(y_pred,y_test):
    print(i)
```

```python
from sklearn.metrics import mean_squared_error, r2_score
```

```python
mean_squared_error(y_test,y_pred)
```

```python
plt.scatter(x_train,y_train,color='red')
plt.plot(x_train,regressor.predict(x_train),'b')

plt.xlabel("RM")
plt.ylabel("TARGET")
plt.title("Linear Regression(train)")
```

```python
plt.scatter(x_test,y_test,color='red')
plt.plot(x_test,regressor.predict(x_test),'b')

plt.xlabel("RM")
plt.ylabel("TARGET")
plt.title("Linear Regression(test)")
```

```python
r2_score(y_test,y_pred)   # tells how much of data present is fit by the model. it tells that if there are 100 pts then 44 pts are fit correctly
```
